```python
################################
#  BANK Account Manager        #
################################
# Description
#     1)there different types of bank account (Checking, Savings, Business)
#     2)Each account will accept deposits and withdrawals, and will need to report balances
##########################

class Account:
    def __init__(self,acct_id,opening_deposit):
        self.acct_id = acct_id
        self.balance = opening_deposit
    

    def __str__(self):  #mehthod to return a recognizable string to any print() command
        return f'${self.balance:.2f}'
    
 
    def deposit(self,dep_amt):  #universal method to accept deposits
        self.balance += dep_amt

    
    def withdraw(self,wd_amt):  #universal method to handle withdrawals
        if self.balance >= wd_amt:
            self.balance -= wd_amt
        else:
            self.balance-=2 #fine of $2 
            print('Funds Unavailable')

class Checking(Account):
    def __init__(self,acct_id,opening_deposit):
        super().__init__(acct_id,opening_deposit) # Run the base class __init__

 
    def __str__(self): #method that returns a string SPECIFIC to Checking accounts
        return f'Checking Account #{self.acct_id}\n  Balance: {Account.__str__(self)}'

class Savings(Account):
    def __init__(self,acct_id,opening_deposit):
        
        super().__init__(acct_id,opening_deposit)  # Run the base class __init__

    
    def __str__(self): #method that returns a string specific to Savings accounts
        return f'Savings Account #{self.acct_id}\n  Balance: {Account.__str__(self)}'


class Business(Account):
    def __init__(self,acct_id,opening_deposit):
        super().__init__(acct_id,opening_deposit) # Run the base class __init__


    def __str__(self): #method that returns a string specific to Business accounts
        return f'Business Account #{self.acct_id}\n  Balance: {Account.__str__(self)}'

class Customer:
    def __init__(self, name, PIN):
        self.name = name
        self.PIN = PIN
        self.accts = {'C':[],'S':[],'B':[]}

    def __str__(self):
        return self.name
        
    def open_checking(self,acct_id,opening_deposit):
        self.accts['C'].append(Checking(acct_id,opening_deposit))
    
    def open_savings(self,acct_id,opening_deposit):
        self.accts['S'].append(Savings(acct_id,opening_deposit))
        
    def open_business(self,acct_id,opening_deposit):
        self.accts['B'].append(Business(acct_id,opening_deposit))
        
    def get_total_deposits(self):
        total = 0
        for acct in self.accts['C']:
            print(acct)
            total += acct.balance
        for acct in self.accts['S']:
            print(acct)
            total += acct.balance
        for acct in self.accts['B']:
            print(acct)
            total += acct.balance
        print(f'Combined Deposits: ${total:.2f}')

def make_dep(cust,acct_type,acct_num,dep_amt):
    """
    make_dep(cust, acct_type, acct_num, dep_amt)
    cust      = variable name (Customer record/ID)
    acct_type = string 'C' 'S' or 'B'
    acct_num  = integer
    dep_amt   = integer
    """
    for acct in cust.accts[acct_type]:
        if acct.acct_id == acct_num:
            acct.deposit(dep_amt)
def make_wd(cust,acct_type,acct_num,wd_amt):
    """
    make_dep(cust, acct_type, acct_num, wd_amt)
    cust      = variable name (Customer record/ID)
    acct_type = string 'C' 'S' or 'B'
    acct_num  = integer
    wd_amt    = integer
    """
    for acct in cust.accts[acct_type]:
        if acct.acct_id == acct_num:
            acct.withdraw(wd_amt)
```
